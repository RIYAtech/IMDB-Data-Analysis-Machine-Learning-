## Classify Movies based on Ratings (Excellent,Good,Avearge)

```python
data.columns
```

```python
def rating(rating):
    if rating>=8.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['rating_category']=data['Rating'].apply(rating)
data.head()
```

## Find Unique values from Genre

```python
data.columns
```

```python
data['Genre']
l=[]
for value in data['Genre']:
    l.append(value.split(','))
l2=[]
for i in l:
    for j in i:
        l2.append(j)
```

```python
l2
```

```python
unique_list=[]
for i in l2:
    if i not in unique_list:
        unique_list.append(i)
```

```python
unique_list
```

```python
len(unique_list)
```
